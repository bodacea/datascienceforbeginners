# Scikit-learn's example datasets

The Python package Scikit-learn (NB you import this as sklearn, not scikit-learn) comes with a collection of example data.  These are all "gold standard": well understood datasets that you can use to test new algorithms, or learn more about how machine learning works.

## Iris

You'll see this dataset a lot in classification examples. It describes 150 plants, all members of the iris family.  What we know about each plant is:
* the width and length of its petals (no, I don't know if that's maximum or average), 
* the width and length of its sepals (that's the green leaf-like part that wraps around a flowerbud)
* its species label (Iris setosa, Iris virginica or Iris versicolor) 

More: https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
from sklearn.datasets import load_iris
irisdata = load_iris()
print(irisdata.keys())
irisdata
```

We can also import and explore some of the example datasets using the Seaborn visualisation library:

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
%matplotlib inline
import seaborn as sns
sns.pairplot(iris, hue='species', size=2)
```

## 20Newsgroups

You'll see this dataset a lot in text processing examples (e.g. text classification). It contains the text from 18,000 newsgroup postings, labelled by the group they come from (rec.autos, sci.space etc; there are 20 groups).

More: 
* http://scikit-learn.org/stable/datasets/twenty_newsgroups.html
* http://qwone.com/~jason/20Newsgroups/

```python
from sklearn.datasets import fetch_20newsgroups
newsgroups_train = fetch_20newsgroups(subset='train')
newsgroups_test = fetch_20newsgroups(subset='test')
print(newsgroups_train.keys())
print(newsgroups_test.keys())
newsgroups_train
```

```python
print('Categories: {}'.format(newsgroups_train['target_names']))
print('\n')
print('First message in dataset: \n{}'.format(newsgroups_train['data'][0]))
```

## Diabetes

The classic test dataset for linear regression, the diabetes dataset contains data for 442 diabetes patients (anonymised, of course). 
* The 10 features in this dataset are: "age, sex, body mass index, average blood pressure, and six blood serum measurements were obtained for each of n = 442 diabetes patients, as well as the response of interest, a quantitative measure of disease progression one year after baseline.". 
* The dataset values in Scikit-learn have been standardised: "normalized to have mean 0 and Euclidean norm 1, and the last column y has been centered (mean 0)."
* The target values represent "a quantitative measure of disease progression one year after baseline". 

More: 
* If you're curious, the original dataset values are here: http://web.stanford.edu/~hastie/Papers/LARS/diabetes.data and are described in Efron et al's paper "Least Angle Regression",  http://web.stanford.edu/~hastie/Papers/LARS/LeastAngle_2002.pdf
* https://archive.ics.uci.edu/ml/datasets/Diabetes

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
print('{}'.format(diabetes.keys()))
diabetes
```

```python
%matplotlib inline
import seaborn as sns
import pandas as pd

diabetesdata = pd.DataFrame(diabetes['data'])
diabetesdata['target'] = diabetes['target']
sns.pairplot(diabetesdata, size=2)
```

```python
# Investigate some of the stranger values above

print('column 1: {}'.format(diabetesdata[1].value_counts()))
print('column 7: {}'.format(diabetesdata[7].value_counts()))
```

## Digits

These are 1700 images of handwritten digits (0 to 9), where each image is 8 pixels wide by 8 pixels tall, and comes with a 64-feature vector (one feature for each pixel, to make training classifiers on this data easier) and a label for the number that the image represents.  The digits dataset is typically used to discuss machine learning on simple image data. 

More:
* http://scikit-learn.org/stable/auto_examples/datasets/plot_digits_last_image.html

```python
from sklearn.datasets import load_digits
digits = load_digits()
print('{}'.format(digits.keys()))
digits
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:10]):
    plt.subplot(2, 5, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('{}'.format(label))
```

Doing a little dimension reduction on the digits dataset, so we can plot it in 2 dimensions instead of 64...  code is from Python for Data Science - see that book for more. 

```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape

plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
edgecolor='none', alpha=0.5,
cmap=plt.cm.get_cmap('spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5)
```

## Boston House Prices

Another classic dataset for testing regression algorithms. Contains 506 houses around Boston, with price and a set of attributes, e.g. 
* CRIM: per-capita crime rate
* PTRATIO: pupil-teacher ratio
* etc

More:
* https://archive.ics.uci.edu/ml/datasets/Housing

```python
from sklearn.datasets import load_boston
boston = load_boston()
print('{}'.format(boston.keys()))
print('{}'.format(boston.data.shape))
boston
```

```python
%matplotlib inline
import seaborn as sns
import pandas as pd

bostondata = pd.DataFrame(boston['data'], columns=boston['feature_names'])
bostondata['target'] = boston['target']
sns.pairplot(bostondata, size=2)
```

```python
print('{}'.format(bostondata['CHAS'].value_counts()))
sns.pairplot(bostondata[['LSTAT', 'target']], size=4)
```
