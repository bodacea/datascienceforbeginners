# R Dataframes

## Factors and Levels

```r
a <- c('developed', 'transitioning', 'developed',
'developing', 'developed', 'developing', 'developed')
classes <- factor(a)
print(classes)
print(levels(classes))
```

## Dataframes

```r
refugees = c(23,175,15543,11,338796,1244,63,53)
asylumseekers = c(140,77,2872,6,255,91,6,7)
origins = c('Burundi','Dem. Rep. of the Congo', 'Eritrea',
'Rwanda', 'Somalia','South Sudan','Sudan','Uganda')
countries <- factor(origins)
popstats <- data.frame(refugees, asylumseekers, countries)
print(popstats)
print(popstats$asylumseekers)
```

## Reading files into Dataframes

list.files() lists all the files in a directory

read.csv reads a comma-separated CSV file into an R dataframe.  

Using read.table with "sep='\t'" reads in a tab-separated CSV file.

```r
list.files()

refugees <- read.csv('../example_data/cleaned_popstats.csv')

population <- read.table('../example_data/sp_pop.tsv', sep='\t', header=TRUE)
```

## Exploring a dataframe

```r
dim(refugees)
str(refugees)
head(refugees)
tail(refugees)
```

## Merging dataframes

Spaces in headings get converted to dots.

Use by.x, by.y *unless* you have at least one column with the same name in both files.

```r
merge(x = refugees, y = population,
by.x='Origin...Returned.from', by.y='Country.Name')
```

## Prettier Visualisations with GGplot2

```r
refugees = c(23,175,15543,11,338796,1244,63,53)
asylumseekers = c(140,77,2872,6,255,91,6,7)
origins = c('Burundi','Dem. Rep. of the Congo', 'Eritrea', 'Rwanda', 'Somalia','South
Sudan','Sudan','Uganda')
countries <- factor(origins)

library(ggplot2)
qplot(refugees, asylumseekers, color = countries)
```
